```python
import pandas as pd
```

```python
df = pd.read_csv('likesfile.csv')
```

```python
amy_likes_set = set(df[df['Name'] == 'Amy'].iloc[:, 1:].values.flatten())
eric_likes_set = set(df[df['Name'] == 'Eric'].iloc[:, 1:].values.flatten())
```

```python
union_result = amy_likes_set.union(eric_likes_set)
print("聯集結果:", union_result)
```

```python
intersection_result = amy_likes_set.intersection(eric_likes_set)
print("交集結果:", intersection_result)
```

```python
difference_result_amy = amy_likes_set.difference(eric_likes_set)
print("差集結果 (Amy相對於Eric的差集):", difference_result_amy)
```

```python
difference_result_eric = eric_likes_set.difference(amy_likes_set)
print("差集結果 (Eric相對於Amy的差集):", difference_result_eric)
```
