```python
import requests
from bs4 import BeautifulSoup
import re

def get_content(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.content
    else:
        print("Failed to fetch content from URL:", url)
        return None

url = "https://www.ntnu.edu.tw/"
content = get_content(url)
if content:
    print("Content fetched successfully!")
else:
    exit()

# 將 content 使用正則表達式清洗資料
cleaned_content = re.sub(r'<[^>]+>', '', content.decode("utf-8"))  # 去除 HTML 標籤

# 測試清洗後的內容
print("Cleaned content:")
print(cleaned_content[:1000])  # 僅列印前 1000 個字元，避免輸出過多
```

```python
import pandas as pd

def structure_content(cleaned_content):
    # 在這裡您可以根據網站內容的結構進行處理，例如提取標題、段落等
    # 這裡只是一個示例，您可以根據實際情況進行修改
    structured_data = {
        "Content": [cleaned_content]  # 這裡將整個內容視為一個項目
    }
    dataframe = pd.DataFrame(structured_data)
    return dataframe, structured_data

dataframe, structured_data = structure_content(cleaned_content)
print("DataFrame:")
print(dataframe.head())

print("\nDictionary:")
print(structured_data)
```

```python
# 輸出 CSV 檔案
dataframe.to_csv("ntnu_web.csv", index=False)

# 輸出 JSON 檔案
import json
with open("ntnu_web.json", "w", encoding="utf-8") as f:
    json.dump(structured_data, f, ensure_ascii=False)

print("CSV and JSON files saved successfully!")
```
