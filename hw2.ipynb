```python
import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt
import seaborn as sns

# read file
csv1 = pd.read_csv('a.csv')
print(csv1)
```

```python
# extract data
season = csv1['season']
pop = csv1['population']

# draw 
plt.figure(figsize=(15, 6))  # set size
plt.plot(season, population, marker='o')  
plt.xlabel('Time')  
plt.ylabel('Population(ten million)')  
plt.xticks(rotation=30)  
plt.show() 
```

```python
# extract data
exchange = csv1['exchange rate']

# draw
plt.figure(figsize=(15, 6))  # set size
plt.plot(season, exchange, marker='o')  
plt.xlabel('Time')  
plt.ylabel('exchange rate')  
plt.xticks(rotation=30)  
plt.show() 
```

```python
# extract data
GDPa = csv1['GDP(average)']
GNIa = csv1['GNI(average)']
incomea = csv1['income(average)']

# draw
plt.figure(figsize=(15, 6))  # set size

plt.plot(season, GDPa, marker='o', label = "GDP")  # gdp
plt.plot(season, GNIa, marker='o', label = "GNI")  # gni
plt.plot(season, incomea, marker='o', label = "INCOME")  # income

plt.title("Economy")  
plt.xlabel('Time')  
plt.ylabel('Dollar')  
leg = plt.legend(loc='lower right')
plt.xticks(rotation=30) 
plt.grid() 
```

```python
# read file
csv2 = pd.read_csv('e.csv')
print(csv2)
```

```python
# extract data
T = csv2["USA's exports to Taipei"]
U = csv2["Taipei's exports to USA"]
year = csv2['year']

# draw
plt.figure(figsize=(15, 6))  # set size
plt.plot(year, T, marker='o', label = "USA's exports to Taipei")  
plt.plot(year, U, marker='o', label = "Taipei's exports to USA")  

plt.title("Exports")  
plt.xlabel('Time')  
plt.ylabel('US Dollar ten billion')  
leg = plt.legend(loc='lower right')
plt.xticks(rotation=30)  
plt.grid()
```

```python
# 
df3 = pd.read_csv('d.csv')
print(df3)
```

```python
# 
rate = df3['s']

# 
labels = ['Primary Industry', 'Secondary Industry', 'Tertiary Industry']
sizes = [rate[0], rate[1], rate[2]]
colors = ['#2f1fde','#e6b815','#8e33e8']
 
fig2, ax2 = plt.subplots()
ax2.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=270)
ax2.axis('equal')

# 
plt.title('stores')

plt.show()
```
