```python
import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt
import seaborn as sns

# read file
csv1 = pd.read_csv('a.csv')
print(csv1)
```

```python
# extract data
season = csv1['season']
pop = csv1['population']

# draw 
plt.figure(figsize=(15, 6))  # set size
plt.plot(season, population, marker='o')  
plt.xlabel('Time')  
plt.ylabel('Population(ten million)')  
plt.xticks(rotation=30)  
plt.show() 
```

```python
# extract data
exchange = csv1['exchange rate']

# draw
plt.figure(figsize=(15, 6))  # set size
plt.plot(season, exchange, marker='o')  
plt.xlabel('Time')  
plt.ylabel('exchange rate')  
plt.xticks(rotation=30)  
plt.show() 
```

```python
# extract data
GDPa = csv1['GDP(average)']
GNIa = csv1['GNI(average)']
incomea = csv1['income(average)']

# draw
plt.figure(figsize=(15, 6))  # set size

plt.plot(season, GDPa, marker='o', label = "GDP")  # gdp
plt.plot(season, GNIa, marker='o', label = "GNI")  # gni
plt.plot(season, incomea, marker='o', label = "INCOME")  # income

plt.title("Economy")  
plt.xlabel('Time')  
plt.ylabel('Dollar')  
leg = plt.legend(loc='lower right')
plt.xticks(rotation=30) 
plt.grid() 
```

```python
'''
是不是人口數量越少，人均所得就會越高?

根據[2]的圖表，人口呈現一個向下的趨勢
根據[4]的圖表，收入雖然沒有一個固定的規律，但是呈現一個向上的趨勢

根據以上兩個資訊，得到人口越少所得雖然不會對所得有絕對的影響，不過還是跟收入提高有關係
'''
```

```python
'''
若探討人口變化的主因
疫情是從108的末期爆發，人口持續減少的主因很有可能是受到疫情影響
因此推論人口變化是跟疫情有關
若探討收入呈現向上趨勢的原因
可能也是因為疫情
高科技產業的需求量變高而造成的結果
-
'''
```

```python
# read file
csv2 = pd.read_csv('e.csv')
print(csv2)
```

```python
# extract data
T = csv2["USA's exports to Taipei"]
U = csv2["Taipei's exports to USA"]
year = csv2['year']

# draw
plt.figure(figsize=(15, 6))  # set size
plt.plot(year, T, marker='o', label = "USA's exports to Taipei")  
plt.plot(year, U, marker='o', label = "Taipei's exports to USA")  

plt.title("Exports")  
plt.xlabel('Time')  
plt.ylabel('US Dollar ten billion')  
leg = plt.legend(loc='lower right')
plt.xticks(rotation=30)  
plt.grid()
```

```python
'''另外，我也發現了一個特別的現象
為什麼台灣的人均收入在第四季時相較於其他季來的高?
-
從資料中可以看出，台灣二三級產業的比例相當高
而每年第四季適逢年終獎金等的發放
所以推測第四季人均收入較高的原因與二三級產業的年終發放有關
-'''
```

```python
# 讀取檔案
df3 = pd.read_csv('d.csv')
print(df3)
```

```python
# 提取資料
rate = df3['s']

# 圓餅圖
labels = ['Primary Industry', 'Secondary Industry', 'Tertiary Industry']
sizes = [rate[0], rate[1], rate[2]]
colors = ['#2f1fde','#e6b815','#8e33e8']
 
fig2, ax2 = plt.subplots()
ax2.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=270)
ax2.axis('equal')

# 添加標題
plt.title('stores')

plt.show()
```

```python
# 提取資料
rate = df3['t']

# 圓餅圖
labels = ['Primary Industry', 'Secondary Industry', 'Tertiary Industry']
sizes = [rate[0], rate[1], rate[2]]
colors = ['#66b3ff','#e6b815','#8e33e8']
 
fig2, ax2 = plt.subplots()
ax2.pie(sizes, labels=labels, colors=colors, autopct='%1.2f%%', startangle=180)
ax2.axis('equal')

# 添加標題
plt.title('turnover(million)')

plt.show()
```
